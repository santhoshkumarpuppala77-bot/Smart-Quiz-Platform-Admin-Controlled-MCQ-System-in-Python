```python
import random

def load_questions():
    questions = []
    
    try:
        with open("questions.txt", "r") as f:
            for line in f:
                parts = line.strip().split("|")
                
                if len(parts) != 6:
                    continue
                
                question = parts[0]
                options = parts[1:5]
                answer = parts[5]
                
                questions.append((question, options, answer))
    
    except FileNotFoundError:
        print("Questions file not found!")
    
    return questions


def signup():
    username = input("Enter new username: ")
    password = input("Enter new password: ")

    try:
        with open("users.txt", "r") as f:
            for line in f:
                u, _ = line.strip().split("|")
                if u == username:
                    print("Username already exists!")
                    return
    except FileNotFoundError:
        pass

    try:
        with open("users.txt", "a") as f:
            f.write(f"{username}|{password}\n")
        print("Signup successful! Please login.")
    except:
        print("Error during signup!")


def login():
    username = input("Enter username: ")
    password = input("Enter password: ")

    try:
        with open("users.txt", "r") as f:
            for line in f:
                u, p = line.strip().split("|")
                if u == username and p == password:
                    print("Login successful!")
                    return username
        
        print("Invalid username or password!")
        return None

    except FileNotFoundError:
        print("No users found! Please signup first.")
        return None


class User:
    def __init__(self, name):
        self.name = name
        self.score = 0

    def take_quiz(self):
        questions = load_questions()
        
        if not questions:
            print("No questions available!")
            return
        
        random.shuffle(questions)
        questions = questions[:10]
        
        for q, options, ans in questions:
            print("\n" + q)
            
            for opt in options:
                print(opt)
            
            while True:
                user_ans = input("Enter option (1-4): ")
                
                if user_ans in ["1", "2", "3", "4"]:
                    break
                else:
                    print("Invalid input! Enter 1-4 only.")
            
            if user_ans == ans:
                print("Correct!")
                self.score += 1
            else:
                print(f"Wrong! Correct answer: {ans}")
        
        print(f"\n{self.name}, Your Score: {self.score}")

    def save_score(self):
        try:
            with open("scores.txt", "a") as f:
                f.write(f"{self.name}|{self.score}\n")
        except:
            print("Error saving score!")


class Admin:
    def add_question(self):
        try:
            with open("questions.txt", "a") as f:
                
                while True:
                    q = input("Enter question: ")
                    
                    options = []
                    for i in range(1, 5):
                        opt = input(f"Enter option {i}: ")
                        options.append(f"{i}) {opt}")
                    
                    while True:
                        answer = input("Enter correct option number (1-4): ")
                        if answer in ["1", "2", "3", "4"]:
                            break
                        else:
                            print("Invalid choice!")
                    
                    f.write(f"{q}|{'|'.join(options)}|{answer}\n")
                    print("Question added successfully!")

                    choice = input("Add another question? (yes/no): ").lower()
                    if choice != "yes":
                        break

        except:
            print("Error saving question!")

    def view_scores(self):
        try:
            with open("scores.txt", "r") as f:
                data = f.readlines()
                
                if not data:
                    print("No scores yet!")
                    return
                
                print("\nLeaderboard:")
                scores = []
                
                for line in data:
                    name, score = line.strip().split("|")
                    scores.append((name, int(score)))
                
                scores.sort(key=lambda x: x[1], reverse=True)
                
                for name, score in scores:
                    print(f"{name} - {score}")
        
        except FileNotFoundError:
            print("No scores file found!")

    def delete_question(self):
        try:
            with open("questions.txt", "r") as f:
                lines = f.readlines()

            if not lines:
                print("No questions available!")
                return

            print("\nAvailable Questions:")
            for i, line in enumerate(lines, 1):
                print(f"{i}. {line.split('|')[0]}")

            choice = int(input("Enter question number to delete: "))

            if 1 <= choice <= len(lines):
                del lines[choice - 1]

                with open("questions.txt", "w") as f:
                    f.writelines(lines)

                print("Question deleted successfully!")
            else:
                print("Invalid choice!")

        except FileNotFoundError:
            print("Questions file not found!")
        except:
            print("Error deleting question!")


def main():
    while True:
        print("\n===== QUIZ SYSTEM =====")
        print("1. User")
        print("2. Admin")
        print("3. Exit")
        
        choice = input("Enter choice: ")
        
        if choice == "1":
            print("\n1. Login")
            print("2. Signup")
            
            auth_choice = input("Enter choice: ")
            
            if auth_choice == "1":
                username = login()
                
                if username:
                    user = User(username)
                    user.take_quiz()
                    user.save_score()
            
            elif auth_choice == "2":
                signup()

        elif choice == "2":
            password = input("Enter admin password: ")
            
            if password == "admin123":
                admin = Admin()
                
                print("\n1. Add Question")
                print("2. View Scores")
                print("3. Delete Question")   
                
                ch = input("Enter choice: ")
                
                if ch == "1":
                    admin.add_question()
                elif ch == "2":
                    admin.view_scores()
                elif ch == "3":
                    admin.delete_question()   
                else:
                    print("Invalid choice!")
            else:
                print("Wrong password!")

        elif choice == "3":
            print("Exiting... Goodbye!")
            break
        
        else:
            print("Invalid option!")


main()
```
